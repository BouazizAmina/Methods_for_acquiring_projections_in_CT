### White gaussian noise and poisson noise

```python
def add_noise(image, peak_poisson, std_gaussian):
    
    # Add Poisson noise
    poisson_noise = np.random.poisson(image * peak_poisson) / peak_poisson
    
    # Add Gaussian noise
    gaussian_noise = np.random.normal(0, std_gaussian, image.shape)
    noisy_image = poisson_noise + gaussian_noise

    # Clip the image between 0 and 255
    noisy_image = np.clip(noisy_image * 255, 0, 255).astype(np.uint8)
    
    # Calcule the std of the noisy image
    std = np.std(noisy_image)
 
    # Transform the image
    maxi, mini = noisy_image.max(), noisy_image.min()
    
    v1 = 2*std
    v2 = maxi - 2*std
    
    a = (v2-v1)/(maxi-mini)
    b = v2-a*maxi
    
    T = a*noisy_image+b

    # Normalize the image
    noisy_image = np.clip(noisy_image/255, 0, 1)
    
    std = np.std(noisy_image)
#     print("std",std*255)
    
#     plt.imshow(noisy_image, cmap = plt.cm.gray)

    return noisy_image
```

```python
from skimage.data import shepp_logan_phantom
from skimage.metrics import structural_similarity as ssim

peak_poisson = 500
std_gaussian = 0.1
i = shepp_logan_phantom()
ii= add_noise(i, peak_poisson, std_gaussian)
plt.imshow(ii, cmap = plt.cm.gray)
print("ssim",ssim(i, ii))
```
